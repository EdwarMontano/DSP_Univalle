# Impulso unitario

```python
# impulso unitario continuo

a = -10
b = 10
dt = 0.01

# procedimento
t = np.arange(a, b, dt)
u0 = np.piecewise(t,t>=0,[1,0])

donde = 0
udonde = np.piecewise(t,t>=donde,[1,0])
udondet = np.piecewise(t,t>=(donde+dt),[1,0])
unimpulso = udonde - udondet


# grafica
plt.figure(1)
plt.plot(t,unimpulso)
plt.margins(dt)
plt.grid()
plt.xlabel('t')
plt.ylabel('impulso()')
plt.title('Impulso continuo ')
plt.show()

# impulso unitario discreto

# procedimento
a1 = -10
dt1 = 1
b1 = 11

t1 = np.arange(a1, b1, dt1)
u0_1 = np.piecewise(t1,t1>=0,[1,0])


donde1 = 0
udonde1 = np.piecewise(t1,t1>=donde1,[1,0])
udondet1 = np.piecewise(t1,t1>=(donde1+dt1),[1,0])
unimpulso1 = udonde1 - udondet1

# grafica
plt.figure(2)
plt.plot(t1,unimpulso1,'o', linewidth=0.1 ,color= 'g')
plt.margins(dt)
plt.grid()
plt.xlabel('t')
plt.ylabel('impulso()')
plt.show()
```

# Función compuerta

```python
# Construcción de la Señal
u2 = np.piecewise(t,t>=2,[1,0])
u4 = np.piecewise(t,t>=4,[1,0])
rectangular = u2 - u4

# SALIDA - grafica
plt.figure(5)
plt.plot(t,rectangular)
plt.xlabel('t')
plt.ylabel('rectangular()')
plt.margins(dt)
plt.show()



# Construcción de la Señal
u2_d = np.piecewise(t1,t1>=2,[1,0])
u4_d = np.piecewise(t1,t1>=6,[1,0])
rectangular_d = u2_d - u4_d

# SALIDA - grafica
plt.figure(6)
plt.plot(t1,rectangular_d,'o', linewidth=0.1 ,color= 'g')
plt.grid()
plt.xlabel('t')
plt.ylabel('rectangular()')
plt.show()
```
